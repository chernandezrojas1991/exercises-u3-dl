```python
import pandas as pd
import numpy as np
```

```python
path = "/content/drive/MyDrive/Coding Dojo/Unsupervised ML/Semana 2/bikeshare_train - bikeshare_train.csv"
df = pd.read_csv(path)
df.head()
```

```python
df.drop(columns=['casual', 'registered'], inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])
df['month_name'] = df['datetime'].dt.strftime('%B')  # Nombre del mes
df['day_of_week'] = df['datetime'].dt.strftime('%A')  # Nombre del día de la semana
df['hour_day'] = df['datetime'].dt.strftime('%H:%M:%S')  # Hora del día
df.head()
```

Ensure month_name, day_of_week and hour_day are object type column

```python
df.info()
```

```python
df.drop(columns=['datetime', 'season'], inplace=True)
df.head()
```

Convert temp and atemp values into Farenheit values

```python
df['atemp'] = df.apply(lambda x: (9/5) * x['atemp'] + 32, axis=1)
df['temp'] = df.apply(lambda x: (9/5) * x['temp'] + 32, axis=1)
df.head()
```

```python
df['temp_variance'] = df['temp'] - df['atemp']
df.drop(columns=['atemp'], inplace=True)
df.head()
```
